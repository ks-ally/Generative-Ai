```python
from transformers import pipeline

# Load GPT-2 model
generator = pipeline(
    'text-generation',
    model='gpt2'
)

# Generate text
result = generator(
    "Artificial Intelligence is",
    max_length=50,
    num_return_sequences=1
)

# Show output
print(result[0]['generated_text'])
```
